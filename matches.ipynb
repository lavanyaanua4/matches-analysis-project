```python
import pandas as pd
df=pd.read_csv("matches (2).csv")
df.head()
df.isnull().sum()
```

```python
df.shape
```

```python
df=df.drop(columns="method")
```

```python
df.isnull().sum()
```

```python
df.describe()
```

Total Matches per season

```python
Matches_per_season=df["season"].value_counts()
```

```python
plt.figure
sns.barplot(x=Matches_per_season.index,y=Matches_per_season.values)
plt.xticks(rotation=75)
plt.title("Total Matches per season")
plt.show()
```

Most Common venues

```python
venue_counts=df["venue"].value_counts().head(10)
```

```python
plt.figure
sns.barplot(x=venue_counts.index,y=venue_counts.values)
plt.xticks(rotation=75)
plt.title("Most Common venues")
plt.show()
```

```python
toss_decision=df["toss_decision"].value_counts(normalize=True)*100
```

```python
plt.figure
plt.pie(df["toss_decision"].value_counts(),labels=df["toss_decision"].value_counts().index,colors=["yellow","green"])
plt.title("Toss distribution")
plt.show()
```

```python
winner_counts=df["winner"].value_counts().head(10)
```

```python
winner_counts
```

```python
plt.figure
sns.barplot(x=winner_counts.index,y=winner_counts.values)
plt.xticks(rotation=75)
plt.title("Total matches won by team")
plt.show()
```

Players of the Match Leaders

```python
player_of_match=df["player_of_match"].value_counts().head(10)
```

```python
player_of_match
```

```python
plt.figure
sns.barplot(x=player_of_match.index,y=player_of_match.values)
plt.xticks(rotation=75)
plt.title("player_of_match top10")
plt.show()
```

Super overs per season

```python
super_over=df.groupby("season")["super_over"].size()
```

```python
plt.figure
sns.barplot(x=super_over.index,y=super_over.values)
plt.xticks(rotation=75)
plt.title("super_over")
plt.show()
```

Result Type Distribution

```python
plt.figure
plt.pie(df["result"].value_counts(),labels=df["result"].value_counts().index)
plt.title("result distribution")
plt.show()
```

Most Avtive Umpires

```python
df = pd.read_csv("matches (2).csv")
df
active_umpires = pd.concat([df["umpire1"],df["umpire2"]]).head()
active_umpires
```

```python
umpires_counts = active_umpires.value_counts()
```

```python
umpires_counts
```

```python
plt.figure
sns.barplot(x=umpires_counts.index,y=umpires_counts.values)
plt.xticks(rotation=75)
plt.title("Most Avtive Umpires")
plt.xlabel("umpires counts")
plt.ylabel("umpires counts")
plt.show()
```
